```python
from sklearn.model_selection import train_test_split
# Train and test sets
X_train, X_test, y_train, y_test = train_test_split(data_df, data.target, test_size=0.20, random_state=0)
# Subtrain and validation sets
X_subtrain, X_val, y_subtrain, y_val = train_test_split(X_train, y_train, test_size=0.20, random_state=0)
```

##Finding the best parameters for `SGDClassifier`:

###Parameter tuning:

```python
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import SGDClassifier

sgd = SGDClassifier(loss='log_loss', random_state = 0)

params = {
    'clf__penalty': ['l2', 'l1', 'elasticnet'],
    'clf__alpha': [0.001, 0.005], #The higher the value, the stronger the regularization.
    'clf__max_iter': [500, 1000],
    'clf__tol': [0.0001, 0.001, 0.01], #The lower the value, the better the generalization.
    'clf__early_stopping': [True],
    'clf__n_iter_no_change': [5, 10] #The lower the value, the better the generalization.
}

pipe = Pipeline([
    ('scale', StandardScaler()),
    ('clf', sgd)
])

grid_search = GridSearchCV(pipe, param_grid = params, cv = 5, scoring = 'roc_auc')
grid_search.fit(X_subtrain, y_subtrain)

print(f"Best Score: {grid_search.best_score_}")
print(f"Best Params: {grid_search.best_params_}")
```

###ROC curve

```python
from sklearn.metrics import roc_curve

y_probs = grid_search.best_estimator_.predict_proba(X_val)[:, 1]
fpr, tpr, thresholds = roc_curve(y_val, y_probs) #y_val is true values, y_probs are predicted probabilities
specificity = 1 - fpr

plt.figure(figsize=(8, 6))

plt.plot(1 - specificity, tpr, marker='o', linestyle='-')
plt.xlabel('1 - Specificity (False Positive Rate)')
plt.ylabel('Sensitivity (True Positive Rate)')
plt.title('Sensitivity vs. 1 - Specificity Curve')
plt.grid(True)

plt.show()
```

###Scoring on a validation set:

```python
from sklearn.metrics import roc_auc_score

best_sgd = grid_search.best_estimator_.named_steps['clf'].set_params(loss='hinge', random_state = 0)

pipe = Pipeline([
    ('scale', StandardScaler()),
    ('clf', best_sgd)
])

pipe.fit(X_subtrain, y_subtrain)
y_val_pred = pipe.predict(X_val)
score = roc_auc_score(y_val, y_val_pred)

print(f"Validation score: {score}")
```

###Scoring on a test set:

```python
y_test_pred = pipe.predict(X_test)
score = roc_auc_score(y_test, y_test_pred)

print(f"Testing score: {score}")
```
